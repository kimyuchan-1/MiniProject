```python
import pandas as pd

df = pd.read_csv("../district_all.csv")

def trim_after_first_space_to_sigungu(name: str) -> str:
    if pd.isna(name):
        return name

    s = str(name).strip()

    # 최소 조건: 공백 1개 이상
    if " " not in s:
        return s

    parts = s.split()
    # parts[0] = 시도(첫 토큰) 가정
    out = [parts[0]]

    found = False
    for tok in parts[1:]:
        out.append(tok)
        if tok.endswith(("시", "군", "구")):
            found = True
            break

    # 시/군/구 토큰을 못 찾으면 원본 유지 (데이터 손실 방지)
    return " ".join(out) if found else s

df["district_name_trimmed"] = df["district_name"].apply(trim_after_first_space_to_sigungu)

df.to_csv("district_all_trimmed.csv", index=False, encoding="utf-8-sig")
```

```python
import pandas as pd

in_path = "../district_all_trimmed.csv"
out_path = "../district_all_trimmed_dedup.csv"

df = pd.read_csv(in_path)

# (district_id, district_name) 동일 조합 중복 제거 (첫 번째만 남김)
df_dedup = df.drop_duplicates(subset=["district_id", "district_name"], keep="first")

# 저장 (엑셀 호환 인코딩)
df_dedup.to_csv(out_path, index=False, encoding="utf-8-sig")
```
