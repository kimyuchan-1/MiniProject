```python
import pandas as pd

# 데이터 로드
df = pd.read_csv("bjd_info_except_boundary.csv", encoding="cp949")

# 조건:
# 문자열 끝이 (시|도) 또는 (시|도 + 공백 + 군|구)
pattern = r'(특별자치도|특별자치시|광역시|특별시|시|도)(\s*(시|군|구))?$'
filtered = df[df['bjd_nm'].str.contains(pattern, regex=True, na=False)]
filtered.to_csv("bjd_filtered_sido_sigungu.csv", index=False, encoding="utf-8-sig")
```

```python
import pandas as pd

df = pd.read_csv("bjd_info_except_boundary.csv", encoding="cp949")

# sgg_nm 존재(공백 제거 후 비어있지 않음)
m = df["sgg_nm"].astype(str).str.strip().ne("") & df["sgg_nm"].notna()
df2 = df[m].copy()

# 원본 행 순서(안정적으로 유지)
df2["_row"] = range(len(df2))

# sgg_nm별 첫 행
out = df2.sort_values("_row").drop_duplicates(subset=["sgg_nm"], keep="first").drop(columns=["_row"])

out.to_csv("bjd_first_by_sgg_nm.csv", index=False, encoding="utf-8-sig")
```
